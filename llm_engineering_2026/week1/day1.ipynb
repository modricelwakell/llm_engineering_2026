```python
# imports

import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI

```

## using openrouter key and check the key 

```python
load_dotenv(override=True)
api_key = os.getenv('OPENROUTER_API_KEY')

if not api_key:
    print("No API key was found - please check your .env file!")
elif not api_key.startswith("sk-or-"):
    print("API key found but it doesn't look like an OpenRouter key (should start with sk-or-)")
else:
    print("API key found and looks good so far!")
```

## calling openai model using api key 

```python
client = OpenAI(
    # this url an endpoint to call openrouter models like GPT
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key
)

message = "hello openai this is my first message for you, hi"

response = client.chat.completions.create(
    model="openai/gpt-5-nano",
    messages=[
        {"role": "user", "content": message}
    ]
)

print(response.choices[0].message.content)
```

differents between user prompt and system prompt

```python
# Define our system prompt - you can experiment with this later, changing the last sentence to 'Respond in markdown in Spanish."

system_prompt = """
system_prompt = 
You are a very skilled website analyst.

Your job:
- Summarize the website in a clear and structured way
- Use Markdown formatting
- Always include:
  - A catchy title
  - Bullet points
  - A short funny or snarky comment at the end
- Make it visually appealing and easy to read

"""
```

```python
# Define our user prompt

user_prompt = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python

response = client.chat.completions.create(
    model="openai/gpt-5-nano",
    messages=[
        {"role": "system", "content": "You are a snarky assistant"},
        {"role": "user", "content": "What is 2 + 2?"}
    ]
)

print(response.choices[0].message.content)
```

```python
def messages_for (website):
    return[
        
    {"role": "system ", "content": system_prompt},
    {"role": "user ", "content": user_prompt+website}
    ]
```

```python
# Try this out, and then try for a few more websites

messages_for(content_website)
```

```python

def summarize(url):
    website = fetch_website_contents(url)

    response = client.chat.completions.create(
        model="openai/gpt-5-nano",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": website}
        ]
    )

    return response.choices[0].message.content
```

```python
summarize("https://edwarddonner.com")
```

using display to show content of website in a beautifull way 

```python
# A function to display this nicely in the output, using markdown

def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://edwarddonner.com")
```

```python
display_summary("https://cnn.com")
```
