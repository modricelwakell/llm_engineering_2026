```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)
api_key = os.getenv('OPENROUTER_API_KEY')

if not api_key:
    print("No API key was found - please check your .env file!")
elif not api_key.startswith("sk-or-"):
    print("API key found but it doesn't look like an OpenRouter key (should start with sk-or-)")
else:
    print("API key found and looks good so far!")
```

### there are two ways to calling API 
1- RAW API(without SDK ) you should write all things by yourself like authorization,URL,CONTET-TYPE

2- with using SDK thats the easiest way because you using model directioanly 
but if you used openrouter you should use base url 

### using SDK

```python
from openai import OpenAI

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key
)

response = client.chat.completions.create(
    model="openai/gpt-5-nano",  # لازم تحدد provider
    messages=[
        {"role": "user", "content": "Tell me a fun fact"}
    ]
)

print(response.choices[0].message.content)
```

### using raw api  without SDK

```python
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}
```

```python

payload = {
    "model": "openai/gpt-5-nano",
    "messages": [
        {"role": "user", "content": "hi, tell me a fun fact"}
    ]
}
```

```python

response = requests.post(
    "https://openrouter.ai/api/v1/chat/completions",
    headers=headers,
    json=payload
)
```

```python

result = response.json()

print(result["choices"][0]["message"]["content"])
```
