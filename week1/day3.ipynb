method of encoding_for_model = callling model

method of encode             = encoding (transform tokens to ID) 

method of decode             = decoding (transform ID to tokens) 

```python
encoding = tiktoken.encoding_for_model('gpt-4.1-mini')
tokens= encoding.encode('Hi my name is Ed and I like banoffee pie')
tokens
```

```python
decode=encoding.decode(tokens)
decode
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENROUTER_API_KEY')

if not api_key:
    print("No API key was found - please check your .env file!")
elif not api_key.startswith("sk-or-"):
    print("API key found but it doesn't look like an OpenRouter key (should start with sk-or-)")
else:
    print("API key found and looks good so far!")
```

```python
from openai import OpenAI
openai = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key
)
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi! I'm Ed!"}
    ]
```

```python
response = openai.chat.completions.create(model="openai/gpt-4.1-mini", messages=messages)
response.choices[0].message.content
```
